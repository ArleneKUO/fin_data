```python
import matplotlib
import matplotlib.pyplot as plt
```

```python
plt.plot(df_merged.index, df_merged['DFF'])
plt.plot(df_merged.index, df_merged['CPIAUCSL'])
```

```python
fig,ax = plt.subplots()
ax.plot(df_merged.index,df_merged['DFF'],color="blue",marker="o")
ax2=ax.twinx()
ax2.plot(df_merged.index, df_merged['CPIAUCSL'],color="red")
fig.savefig('折線圖.jpg',format='jpeg')
```
